```python
import pandas as pd
import xarray as xr
import netCDF4 as nc
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
```

```python
data= xr.open_mfdataset('*.nc')
data
```

```python
data=data.where(data !=-99.0)
```

```python
data
```

```python
japan_rr=data.precip
japan_rr
```

```python
japan_RR=japan_rr.sel(lon=138.25, lat=36.20, method='nearest')
```

```python
japan_RR
```

```python
monthlyrr_total=japan_RR.resample(datetime='1M').sum('datetime')
monthlyrr_total
```

```python
monththly_climo=monthlyrr_total.groupby('datetime.month').mean('datetime')
monththly_climo
```

```python
annualrr_total=japan_RR.resample(datetime='1Y').sum('datetime')
annualrr_total
```

```python
annual_total=japan_rr.groupby('datetime.year').sum()
annual_total

annual_clime=annual_total.mean(dim=['lat','lon'])
annual_clime
```

# SPATIAL PLOTS

```python
dryDays_per_year_ts = japan_rr.where(japan_rr < 1).groupby('datetime.year').count('datetime')
wetDays_per_year_ts = japan_rr.where(japan_rr >= 1).groupby('datetime.year').count('datetime')
dryDays_per_month_ts = japan_rr.where(japan_rr < 1).groupby('datetime.month').count('datetime')
wetDays_per_month_ts = japan_rr.where(japan_rr >= 1).groupby('datetime.month').count('datetime')
```

```python
#spatial plots for dry days per year
fig,ax=plt.subplots(2,5,figsize=(15,5), 
                    subplot_kw={'projection': ccrs.PlateCarree()})
ax=ax.flatten()
year_names=['2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']
for i in range(10):
    ax[i].add_feature(cfeature.COASTLINE,linewidth=0.5)
    ax[i].add_feature(cfeature.BORDERS,linewidth=2)
    ax[i].add_feature(cfeature.STATES,linewidth=0.5)
     # ax[i].add_feature(cfeature.OCEAN,linewidth=2)
       # ax[i].add_feature(cfeature.LAKES,linewidth=2 color='blue')
        #    ax[i].add_feature(cfeature.RIVERS)
    ax[i].set_extent([125,152,30,45])
        
    ax[i].set_title(year_names[i])
    cb=ax[i].contourf(dryDays_per_year_ts.lon,dryDays_per_year_ts.lat,dryDays_per_year_ts[i],cmap='coolwarm',transform=ccrs.PlateCarree())
    color_bar=fig.add_axes([0.92,0.29,0.025,0.5])
fig.colorbar(cb,cax=color_bar,label='precipitation counts')
plt.subplots_adjust(wspace=-0.10,top=0.9)
plt.suptitle('spatial plots for dry days per year',fontweight='bold')
plt.savefig('dry days per year.png')
```
